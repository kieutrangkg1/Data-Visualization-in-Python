**Pyplot API với Object_Oriented API **
* Pyplot API --> Quickly
* 00 API --> Anvanced

**Pyplot API**

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'dashed', label ='sin(x)');
plt.plot(x,np.cos(x), color = 'red', label = 'cos(x)')  # vẽ thêm một plot
plt.title('Sin(X) and Cos(X)')  # tên của đồ thị muốn vẽ
plt.xlabel('X')
plt.ylabel('F(x)')
# plt.xlim([0,4])  # giới hạn từ 0 đến 4
# plt.ylim([-0.7, 0.6])
# plt.axis([xmin, xmax, ymin, ymax])  # dùng thay vì lim
plt.axis([0,4,-0.7,0.6])
# plt.axis("tight") # nếu muốn cái đó fit
plt.axis("equal")
plt.legend(); #nếu muốn label thì dùng legend nó mới label được
```

**Objected-Oriented API**

```python
# 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

#2. Setup plot
fig, ax = plt.subplots(figsize= (5,5))  #figure size = width and height of the plot
# ax là từng hình nhỏ, fig là hình tổng

# 3. Plot the data
ax.plot(x,y);
ax.set(title = 'A simple plot',
           xlabel = 'x-axis',
           ylabel = 'y-axis')
```

**Scatter **

```python
# Pyplot API
plt.scatter(x, np.exp(x)) # y = e^x
```

```python
# OO API

# Prepare New Data
rng = np.random.RandomState(0)
x = rng.randn(100)  # randn sẽ return lấy ra từ standard normal distribution
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)  # kích cỡ của mỗi cặp

fig, ax = plt.subplots()
img1 = ax.scatter(x,y, s =sizes, c = colors, cmap = 'viridis', alpha = 0.3 )  # cmap: chọn màu của những chấm tròn theo team mà mình mong muốn
                                                                                                         # alpha: chỉnh độ trong suốt

fig.colorbar(img1)   # với màu nàu tương ứng với mức độ bao nhiêu
```

**Bar**
* Vertical
* Horizontal

```python
# Prepare Data from a python dictionary
soft_drink_prices = {'Coke': 10,
                                   'Pessi': 12,
                                   'Sprite': 8}
fig, ax = plt.subplots()
ax.bar(soft_drink_prices.keys(), soft_drink_prices.values())
ax.set(title = "Back Hoa Xanh's Soft Drink Prices", ylabel = "Price ($)")
```

```python
fig, ax = plt.subplots()
ax.barh(list(soft_drink_prices.keys()), list(soft_drink_prices.values()))  # bar horinzontal không nhận được giá trị từ dictionary nên phải chuyển thành list
```

**Histogram**

```python
# Prepare Data
np.random.seed(42)  #
student_height = np.random.normal(170, 10, 250) # min, độ lệch chuẩn, max
student_height[:10], len(student_height) # bao nhiêu bạn trên m7
```

```python
# pyplot API
plt.hist(student_height, bins = 30) # bin là để chia nhỏ ra nữa
```

**Subplots trên cùng 1 figure **

```python
# Option 1:
fig, ((ax1, ax2), (ax3, ax4))  = plt.subplots(nrows = 2,
                                  ncols = 2,
                                  figsize = (10,5))
# Plot data to each axis
ax1.plot(x, x/2);
ax1.plot(x,x*2);

ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax4.hist(student_height, bins = 30);
```

```python
# Option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))
# Plot data to each axis
ax[0,0].plot(x, x/2);
ax[0,0].plot(x,x*2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(), soft_drink_prices.values());
ax[1,1].hist(student_height, bins = 30);
```

**Project: Data Visualization - California Cities **

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('/content/drive/MyDrive/Python của trang/Python Nâng Cao/california_cities.csv')
cities.head()
cities["area_total_km2"].sort_values(ascending = False)
# latd: vĩ độ
# longd: kinh độ
```

```python
# Extract Latd and Longd
lat, lon = cities["latd"], cities["longd"]
lat[:10]
population, area = cities['population_total'], cities['area_total_km2']
plt.style.use('seaborn')
plt.figure(figsize = (8,6))
# Plot using Pyplot API
plt.scatter(lon, lat, c= np.log10(population), cmap = 'viridis', # hiện ra vĩ độ
                  s = area, linewidths = 0, alpha = 0.5 )   # tùy chỉnh kích cỡ, để biết thành phố nào có diện tích lớn bé như thế nào, tức là từng chấm có area khác nhau
plt.axis('equal') # trang trí cho bằng nhau, đều hơn
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label = 'log$_{10}$(population)') # đây là cách gõ log
plt.clim(3,7) # thu hẹp lại cái colorbar scale

# thêm chú thích ở các chấm ( Create a legend for city size)
area_range = [50,100,300,500]
for area in area_range:
      plt.scatter([], [], s= area, c = 'k', alpha = 0.4, label = str(area)+ 'km$^2' )
plt.legend(labelspacing = 1, title = 'City Area ')
plt.title('California Cities: Population and Area Distribution')
# thêm chú thích ở các chấm
```
